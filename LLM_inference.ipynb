```python
model_id = "TinyLlama/TinyLlama-1.1B-Chat-v1.0"
local_dir = "models/TinyLlama-1.1B-Chat-v1.0"
# we chose tiny lama because small data (only 1Billion + no access code required)
```

```python
from huggingface_hub import snapshot_download
snapshot_download(
    repo_id=model_id,
    local_dir=local_dir,
    local_dir_use_symlinks=False,
)
print("Downloaded to:", local_dir)
```

```python
from transformers import AutoTokenizer, AutoModelForCausalLM
import torch
tokenizer = AutoTokenizer.from_pretrained(local_dir, local_files_only=True)
device = "cuda" if torch.cuda.is_available() else "cpu"
# autotokenizers transforms things into tokens
# tokens splits things into small parts e.g Wh/at /is /G/D/P and assigns numbers to them
```

```python
model = AutoModelForCausalLM.from_pretrained(
local_dir,
torch_dtype=torch.float32, # or bfloat16/float16 if GPU supports it
local_files_only=True, # prevents accidental internet access
).to(device)
```

```python

model = AutoModelForCausalLM.from_pretrained(
    local_dir,
    local_files_only =True,
    torch_dtype = torch.float32
).to(device)
```

```python
messages = [
    {"role":"system", "content":"You are a helpful assistant"},
    {"role":"user", "content":"What is GDP in economy?"}
]
```

```python
inputs = tokenizer.apply_chat_template(
    messages,
    add_generation_prompt=True,
    tokenize=True,
    return_dict=True,
    return_tensors="pt"
).to(device)
```

```python
outputs = model.generate(**inputs,max_new_tokens=80)
```
